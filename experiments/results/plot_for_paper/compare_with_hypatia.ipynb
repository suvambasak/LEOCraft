# Comparing the LEOCraft with Hypatia

Hypatia [1][2] is one of the most well known packet level LEO satellite network simulation framework. 
We compared the computed RTT and simulation run time of LEOCraft [4] with Hypatia [3] ping measurement and same simulation run time.

**Reference**

1. Kassing S, Bhattacherjee D, √Åguas AB, Saethre JE, Singla A. Exploring the" Internet from space" with Hypatia. InProceedings of the ACM Internet Measurement conference 2020 Oct 27 (pp. 214-229).
2. https://github.com/snkas/hypatia.git
3. https://github.com/suvambasak/hypatia/tree/master/evaluation
4. https://github.com/suvambasak/LEOCraft/tree/main/experiments/evaluation

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# plt.rcParams["figure.figsize"] = (20, 7)
# plt.rcParams.update({'font.size': 20})
plt.rcParams["figure.figsize"] = (21, 9)
plt.rcParams["figure.figsize"] = (17, 9)
plt.rcParams.update({'font.size': 45})
```

## Compare computed RTT Measument of LEOCraft with ping in Hypatia

```python
def prepare_csv(csv_file):
    df = pd.read_csv(csv_file)
    df['time_s'] = df['time_ns']/1000000000
    df['RTT_ms'] = df['ping_ns']/1000000
    return df
```

### With Kuiper constellation

```python
h_ping_delhi_to_new_york_rtt_df = prepare_csv(
    'CSVs/compare_simulation_platform/Hypatia/kuiper_k1/ping_delhi_to_new-york_rtt.csv')
h_ping_moscow_to_paris_rtt_df = prepare_csv(
    'CSVs/compare_simulation_platform/Hypatia/kuiper_k1/ping_moscow_to_paris_rtt.csv')
h_ping_tokyo_to_sydney_rtt_df = prepare_csv(
    'CSVs/compare_simulation_platform/Hypatia/kuiper_k1/ping_tokyo_to_sydney_rtt.csv')

l_ping_delhi_to_new_york_rtt_df = pd.read_csv(
    'CSVs/compare_simulation_platform/LEOCraft/kuiper_k1/ping_delhi_to_new-york_rtt.csv')
l_ping_moscow_to_paris_rtt_df = pd.read_csv(
    'CSVs/compare_simulation_platform/LEOCraft/kuiper_k1/ping_moscow_to_paris_rtt.csv')
l_ping_tokyo_to_sydney_rtt_df = pd.read_csv(
    'CSVs/compare_simulation_platform/LEOCraft/kuiper_k1/ping_tokyo_to_sydney_rtt.csv')


fig, ax1 = plt.subplots()
ax1.set_xlabel('Time (s)')
ax1.set_ylabel('RTT (ms)')


# LEOcraft
ax1.plot(l_ping_delhi_to_new_york_rtt_df['time_s'], l_ping_delhi_to_new_york_rtt_df['RTT_ms'],
         label='[LEOcraft] New Delhi - New York', color='g', linewidth=6)
ax1.plot(l_ping_moscow_to_paris_rtt_df['time_s'], l_ping_moscow_to_paris_rtt_df['RTT_ms'],
         label='[LEOcraft] Moscow - Paris', color='r', linewidth=6)
ax1.plot(l_ping_tokyo_to_sydney_rtt_df['time_s'], l_ping_tokyo_to_sydney_rtt_df['RTT_ms'],
         label='[LEOcraft] Tokyo - Sydney', color='b', linewidth=6)


# Hypatia
ax1.plot(h_ping_delhi_to_new_york_rtt_df['time_s'], h_ping_delhi_to_new_york_rtt_df['RTT_ms'],
         label='[Hypatia] New Delhi - New York', color='g', linewidth=13, linestyle='dotted')
ax1.plot(h_ping_moscow_to_paris_rtt_df['time_s'], h_ping_moscow_to_paris_rtt_df['RTT_ms'],
         label='[Hypatia] Moscow - Paris', color='r', linewidth=13, linestyle='dotted')
ax1.plot(h_ping_tokyo_to_sydney_rtt_df['time_s'], h_ping_tokyo_to_sydney_rtt_df['RTT_ms'],
         label='[Hypatia] Tokyo - Sydney', color='b', linewidth=13, linestyle='dotted')


ax1.set_yticks([0, 50, 100, 150, 200])
# ax1.set_ylim(-1, 210)
ax1.set_ylim(-1, 160)
# ax1.legend(framealpha=0, prop={'size': 35}, ncol=2)
# ax1.legend(framealpha=0.5, prop={'size': 31.5}, ncol=2)
plt.tight_layout()
plt.grid()
plt.show()
```

## Runtime comparison of Hypatia and LEOCraft

### Regular

```python
plt.rcParams["figure.figsize"] = (27, 9)

hypatia_runime_df = pd.read_csv(
    'CSVs/compare_simulation_platform/RunTime/hypatiaRunTime.csv')
hypatia_runime_df['sat'] = hypatia_runime_df['o']*hypatia_runime_df['n']
hypatia_runime_df['time_s'] = hypatia_runime_df['time_s']/(60*60)

leocraft_runime_df = pd.read_csv(
    'CSVs/compare_simulation_platform/RunTime/LEOCraftRunTime.csv')
leocraft_runime_df['sat'] = leocraft_runime_df['o']*leocraft_runime_df['n']
leocraft_runime_df['time_s'] = leocraft_runime_df['time_s']/(60*60)

x = np.arange(len(leocraft_runime_df['sat']))
width = 0.4

plt.bar(x-0.2, leocraft_runime_df['time_s'], width,
        hatch='|', color='lightgreen', label='LEOCraft')
plt.bar(x+0.2, hypatia_runime_df['time_s'], width,
        hatch='/', color='orange', label='Hypatia')


plt.xticks(rotation=23)
plt.xticks(x, [f"{o}x{o}" for o in leocraft_runime_df['o']])
plt.xlabel("# Satellites")
plt.ylabel("Time (hours)")
plt.legend()
plt.grid()
plt.show()
```

### Time in Log scale

```python
plt.rcParams["figure.figsize"] = (27, 9)

hypatia_runime_df = pd.read_csv(
    'CSVs/compare_simulation_platform/RunTime/hypatiaRunTime.csv')
hypatia_runime_df['sat'] = hypatia_runime_df['o']*hypatia_runime_df['n']
hypatia_runime_df['time_s'] = hypatia_runime_df['time_s']/(60*60)

leocraft_runime_df = pd.read_csv(
    'CSVs/compare_simulation_platform/RunTime/LEOCraftRunTime.csv')
leocraft_runime_df['sat'] = leocraft_runime_df['o']*leocraft_runime_df['n']
leocraft_runime_df['time_s'] = leocraft_runime_df['time_s']/(60*60)

x = np.arange(len(leocraft_runime_df['sat']))
width = 0.4

plt.bar(x-0.2, leocraft_runime_df['time_s'], width,
        hatch='|', color='lightgreen', label='LEOCraft')
plt.bar(x+0.2, hypatia_runime_df['time_s'], width,
        hatch='/', color='orange', label='Hypatia')

plt.yscale("log")
plt.xticks(rotation=23)
plt.xticks(x, [f"{o}x{o}" for o in leocraft_runime_df['o']])
plt.xlabel("# Satellites")
plt.ylabel("Time (hours)")
plt.legend()
plt.grid()
plt.show()
```
