## LEO constellation size validation

```python
import pandas as pd
import matplotlib.pyplot as plt
# plt.rcParams["figure.figsize"] = (20, 7)
# plt.rcParams.update({'font.size': 20})
plt.rcParams["figure.figsize"] = (22, 10)
plt.rcParams.update({'font.size': 45})
```

Checking minimum budget of satellites in a LEO constellation for exhibit behaviour $0 >> n$ at $p=50$ gives higher throughput.

```python
df = pd.read_csv('CSVs/constellation_size_validation/120.csv')
_o = [10, 12]


for i, o in enumerate(_o):
    df_by_o = df.loc[df['S0_o'] == o].sort_values(by='S0_p')
    plt.plot(df_by_o['S0_p'], df_by_o['throughput_Gbps'],
             label=f'o={o}', linewidth=i+1)

plt.title('120')
plt.xlabel('Phase-offset (p) %')
plt.ylabel('Throughput (Tbps)')
plt.legend(ncol=2)
plt.show()
```

### Performance fluctuation with budget

```python
for f in [
    'CSVs/constellation_size_validation/30x30.csv',
    'CSVs/constellation_size_validation/40x40.csv',
    'CSVs/constellation_size_validation/50x50.csv',
    'CSVs/constellation_size_validation/60x60.csv'
]:
    df = pd.read_csv(f).sort_values(by='time_delta')
    df['time_delta'] = df['time_delta'] * 24

    print(
        'var: ',
        df['throughput_Gbps'].var(),
        'STD: ',
        df['throughput_Gbps'].std()
    )


for f in [
    'CSVs/constellation_size_validation/30x30.csv',
    'CSVs/constellation_size_validation/40x40.csv',
    'CSVs/constellation_size_validation/50x50.csv',
    'CSVs/constellation_size_validation/60x60.csv'
]:
    df = pd.read_csv(f).sort_values(by='time_delta')
    df['time_delta'] = df['time_delta'] * 24

    x_line = [h for h in range(0, 25, 3)]
    fig, ax1 = plt.subplots()
    ax1.set_xticks(x_line, minor=False)
    ax1.xaxis.grid(True, which='major')
    ax1.xaxis.grid(True, which='minor')
    ax1.set_xlabel('Time (hours)')
    ax1.set_ylabel('Throughput (Gbps)')
    ax1.tick_params(axis='y')
    ax1.plot(df['time_delta'], df['throughput_Gbps'], color='r', linewidth=4)
    ax1.tick_params(axis='y')
    fig.tight_layout()
    # ax1.legend()
    plt.tight_layout()
    # plt.savefig()
    plt.show()
```

### Performance fluctuation with budget when design optimized

```python
for f in [
    'CSVs/constellation_size_validation/30x30_optimized.csv',
    'CSVs/constellation_size_validation/40x40_optimized.csv',
    'CSVs/constellation_size_validation/50x50_optimized.csv',
    'CSVs/constellation_size_validation/60x60_optimized.csv'
]:
    df = pd.read_csv(f).sort_values(by='time_delta')
    df['time_delta'] = df['time_delta'] * 24

    print('______________________________________________________')
    print('VAR:', df['throughput_Gbps'].var())
    print('STD:', df['throughput_Gbps'].std())
    print('MEAN:', df['throughput_Gbps'].mean())
    print('MEDIAN:', df['throughput_Gbps'].median())
    print('MAX:', df['throughput_Gbps'].max())
    print('MIN:', df['throughput_Gbps'].min())


for f in [
    'CSVs/constellation_size_validation/30x30_optimized.csv',
    'CSVs/constellation_size_validation/40x40_optimized.csv',
    'CSVs/constellation_size_validation/50x50_optimized.csv',
    'CSVs/constellation_size_validation/60x60_optimized.csv'
]:
    df = pd.read_csv(f).sort_values(by='time_delta')
    df['time_delta'] = df['time_delta'] * 24

    x_line = [h for h in range(0, 25, 3)]
    fig, ax1 = plt.subplots()
    ax1.set_xticks(x_line, minor=False)
    ax1.xaxis.grid(True, which='major')
    ax1.xaxis.grid(True, which='minor')
    ax1.set_xlabel('Time (hours)')
    ax1.set_ylabel('Throughput (Gbps)')
    ax1.tick_params(axis='y')
    ax1.plot(df['time_delta'], df['throughput_Gbps'], color='r', linewidth=4)
    ax1.tick_params(axis='y')
    fig.tight_layout()
    # ax1.legend()
    # plt.tight_layout()
    # plt.savefig()
    plt.show()
```
